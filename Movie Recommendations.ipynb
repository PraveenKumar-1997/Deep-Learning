```python
#https://medium.com/code-heroku/building-a-movie-recommendation-engine-in-python-using-scikit-learn-c7489d7cb145
import pandas as pd
import numpy as np
import os
os.chdir('C:\\Users\\ABIARYAA\\Documents')
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movie_dataset.csv")
```

```python
df.index
```

```python
df.tail()
```

```python
def combine_features(row):
    return str(row['keywords']) +" "+str(row['cast'])+" "+str(row["genres"])+" "+str(row["director"])

df["combined_features"] = df.apply(combine_features, axis = 1)
df['combined_features'].shape
```

```python
features = ['keywords','cast','genres','director']
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
cv = CountVectorizer() #represent texts as vectors
count_matrix = cv.fit_transform(df["combined_features"])
count_matrix
```

```python
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim.shape)
print(cosine_sim)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
movie_user_likes = "Avatar"
movie_index = get_index_from_title(movie_user_likes)
movie_index
```

```python
similar_movies =  list(enumerate(cosine_sim[movie_index]))
print(len(similar_movies))
print(similar_movies)
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
print(sorted_similar_movies)
```

```python
i=0
print("Top 5 similar movies to " + movie_user_likes + " are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>=5:
        break
```
